## Bibliotecas Usadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.model_selection import learning_curve
from sklearn.metrics import roc_curve, auc, plot_confusion_matrix
                                              
# Classifiers
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

## Processamento e leitura dos dados
No bloco a seguir utilizamos a biblioteca pandas para ler os valores do CSV. Após a leitura, assinalamos à y a coluna referente ao resultado da classificação e substituímos "CONFIRMED" e "FALSE POSITIVE" por 1 e 0. Assinalamos a X os valores das outras colunas e a partir dos valores de X e y criamos os conjuntos de teste e de treinamento.

```python
df = pd.read_csv("koi_data.csv")

y = df["koi_disposition"].replace({ 'CONFIRMED': 1, 'FALSE POSITIVE': 0 })


# Remove koi_disposition and koi_name from features
X = df.iloc[:, 2:-1]
print("Tamanho de X: ", X.shape)
print("Tamanho de y: ", y.shape)
print("Número de dados - M = ", X.shape[0])
print("Número de features - N = ", X.shape[1])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

## Funções de utilidade e gráficos

A função do bloco abaixo, recebe um classificador e exibe o gráfico da curva de aprendizado para valores diferentes do set de treinamento. As curvas de aprendizado são importantes para reconhecer se a precisão do modelo aumenta a medida que aumentamos sua capacidade. À partir desta curva, conseguimos diagnosticar se o modelo tem consegue generalizar bem ou se ele está propenso a overfit.

```python
# based on https://chrisalbo n.com/machine_learning/model_evaluation/plot_the_learning_curve/
def plotLearningCurve(clf, ax_plt):
        train_size = [500, 1000, 2500, 4000]
        train_sizes, train_scores, test_scores = \
            learning_curve(clf, X, y, cv=5, shuffle=True, random_state=42, \
                           scoring='accuracy', train_sizes=train_size)
        
        train_mean = np.mean(train_scores, axis=1)
        test_mean = np.mean(test_scores, axis=1)

        ax_plt.plot(train_sizes, train_mean, '--', color="#961515",  label="Training score")
        ax_plt.plot(train_sizes, test_mean, color="#2135a6", label="Cross-validation score")

        ax_plt.set_xlabel("Training Set Size"), ax_plt.set_ylabel("Accuracy Score"), ax_plt.legend(loc="best")
        
```

A função encontrada no bloco a seguir recebe uma lista de classificadores e renderiza um gráfico com as curvas ROC de cada um deles. A curva ROC tem a utilidade de mostrar o quão acertivo o modelo é, através da relação entre TPR e FPR. O True Positive Rate recebe todas a saídas positivas do modelo (1 ou "CONFIRMED") e mede em quantas dessas ele acertou, já o False Positive Rate analisa todos os erros produzidos e avalia quantos deles vieram de False Positives. \
O valor da Area Under Curve (AUC) descreve a capacidade do modelo de prever as classes corretamente, ou seja quanto mais alto o valor da AUC, maior será a taxa de acerto do modelo em prever "FALSE POSITIVE" como "FALSE POSITIVE" e "CONFIRMED" como "CONFIRMED".

```python
# based on https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html#sphx-glr-auto-examples-model-selection-plot-roc-py
def plotROCCurves(classifiers):
    plt.figure()
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title('ROC curves')
    for name, clf, _ in classifiers:
        if hasattr(clf, "decision_function"):
            y_score = clf.decision_function(X_test)
        else:
            y_score = clf.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_score)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label='%s (AUC area = %0.2f)' % (name, roc_auc))
        
    plt.legend(loc="lower right")
    plt.show()
    
```

A função a seguir recebe uma lista de classificadores e exibe um grid de classificadores, no qual para cada classificador é exibido a curva de aprendizado e a matriz de confusão. Além disso, é possível observar que a acurácia esperada (obtida no test_set) é exibida ao lado do título em parênteses.

---
A matriz de confusão é composta por 2 eixos que formam quatro quadrantes, no qual se mede a porcentagem de predição entre true positives, true negatives, false positives e false negatives. À partir da interpretação dos quadrantes, conseguimos avaliar a acertividade do modelo a partir da interpretação dos acertos e erros em cada classe.

```python
def plotClassifierCharts(classifiers):
    if len(classifiers) == 1:
        plt.figure(figsize=(6, 6))
    else:
        plt.figure(figsize=(18, 6))
        
    i=1
    for name, clf, score in classifiers:
        ax = plt.subplot(2, len(classifiers), i)
        ax.set_title("%s, (ac: %.4f)" % (name, score))
        plotLearningCurve(clf, ax)
        i+=1
    
    for name, clf, _ in classifiers:
        ax = plt.subplot(2, len(classifiers), i)
        plot_confusion_matrix(clf, X_test, y_test, normalize='true', cmap=plt.cm.Blues, ax=ax)
        i+=1
        
    plt.tight_layout()
    plt.show()    
```

## Treino e avaliação

Nos seguintes blocos notamos seguintes funções: 
- `KFoldTrain`: Recebe um classificador e o conjunto de treinamento e executa o estágio de treinamento no modelo utilizando k-fold com k = 5;
- `trainAndEvaluate`: Recebe uma lista de classificadores e executa o KFoldTrain em cada um, após isso ele executa as funções que exibem os gráficos.
---
Na função `trainAndEvaluate`, o classificador é criado a partir de uma pipeline, com intuíto de impedir que haja memory leak entre as etapas. O pipeline possui os seguintes passos:
- scaler: Executado pelo `StandardScaler()`, essa etapa aplica o mean normalization, que faz com que as entradas das features possuam valores contidos em um intervalo [a,b], isso ajuda o algoritmo a encontrar o mínimo local dando passos mais regulares, já que os valores das features estão mais próximos.
- classification: Essa etapa simplesmente recebe o classificador e o executa no final da execução da pipeline.

```python
def KFoldTrain(X_train_total, y_train_total, clf):
    scores = []
    cv = KFold(n_splits=5, shuffle=True)
    
    for train_index, test_index in cv.split(X_train_total):
        X_train_p, X_test_p, y_train_p, y_test_p = \
            X_train_total[train_index], X_train_total[test_index], \
            y_train_total[train_index], y_train_total[test_index]
        
        clf.fit(X_train_p, y_train_p)
        scores.append(clf.score(X_test_p, y_test_p))
        
    return scores
```

```python
def trainAndEvaluate(classifiers):
    trained_clfs = []
    for name, clf in classifiers:
        pipe_clf = Pipeline([
            ('scaler', StandardScaler()),
            ('classification', clf)
        ])
        
        KFoldTrain(X_train.values, y_train.values, pipe_clf)
        score = pipe_clf.score(X_test, y_test)
        trained_clfs.append([name, pipe_clf, score])
        
    plotClassifierCharts(trained_clfs)          
    plotROCCurves(trained_clfs)
```

## SVM
A principal característica de um classificador que utiliza Suport Vector Machine (SVM) é a utilização de uma margem entre o classificador e os pontos. Essa margem permite que o algoritmo posicione o separador de forma ótima - com a mesma distância entre os pontos da classe negativa e positiva. \
O algoritmo de aprendizado do SVM posiciona o separador de forma a minimizar o erro e maximizar a margem de segurança. A margem de segurança é parametrizada pelo hiperparâmetro C, que age como um regularizador que penaliza a maximização do distância da margem e os dados - quanto maior o valor de C, menor a margem de segurança.

```python
svm_classifiers = [
    ['Linear Kernel', SVC(kernel="linear", C=1)],
    ['Sigmoid Kernel', SVC(kernel="sigmoid", C=1)],
    ['Polinomial Kernel', SVC(kernel="poly", C=1)],
    ['RBF Kernel', SVC(kernel="rbf", C=1)]
];

trainAndEvaluate(svm_classifiers)
```

Nesse experimento, testamos a precisão dos algoritmos de SVM, a partir da variação dos Kernels a seguir:
- **Linear Kernel**: O linear kernel, mantém a estrutura dos dados com o formato original e busca criar o hiperplano que melhor separa os dados. Podemos perceber que, dentre os experimentos, este foi o que apresentou os melhores resultados tanto em acurácia no conjunto de teste quanto em termos de Precision (abservado na matriz de confusão) e Recall (observado no eixo True positive rate na curva ROC). O fato deste kernel ter performado melhor indica que o dataset está disposto linearmente separável
- **Sigmoid Kernel**: Esse foi o kernel com a pior performance em relação aos outros, à partir da curva de aprendizado, conseguimos perceber que a medida que a Capacidade aumenta, o algoritmo performa de forma pior, indicando uma incapacidade de aprender mais com novos dados, o que pode ser interpretado como Underfit. À partir da matriz de confusão, percebemos que o algoritmo teve sua pior performance em classificar valores positivos, acertando apenas 79% das previsões. Na curva ROC, percebemos que isso é reinforçado pelo desempenho no eixo True Positive Rate e pela área da Curva AUC.
- **Polinomial Kernel**: À partir de sua matriz de confusão, percebemos que o kernel polinomial obteve um ótimo resultado em predições de valores positivos ("CONFIRMED"), obtendo uma precisão de 97%. Porém a performance geral foi afetada pela menor precisão em valores negativos. Observando a curva de aprendizado, percebemos que 
- **RBF Kernel**: O kernel RBF teve a segunda melhor performance, obtendo 92.07% de precisão. Além disso ele apresentou um alto score de Recall (.94) e Precision (.92).

## Random Forest
O random forest é um classificador composto por N Decision Trees que retornam valores distintos, e em uma classificação, o random forest seleciona o resultado da Decision Tree que possuir o maior score.

```python
rf_classifiers = [
    ['1 trees RF', RandomForestClassifier(max_depth=5, n_estimators=1, max_features=5)],
    ['10 trees RF', RandomForestClassifier(max_depth=5, n_estimators=10, max_features=5)],
    ['50 trees RF', RandomForestClassifier(max_depth=5, n_estimators=50, max_features=5)],
    ['100 trees RF', RandomForestClassifier(max_depth=5, n_estimators=100, max_features=5)],
]

trainAndEvaluate(rf_classifiers)
```

Para avaliar a performance do algoritmo de Random Forest, quatro experimentos foram conduzidos, variando o número de estimadores (Decision Trees) de cada classificador. Para cada classificador, utilizamos o seguinte número de estimadores:
- **1 Decision Tree**: Ao analisar a curva de aprendizado deste modelo, observamos que quando aumentamos a capacidade do modelo para além de 2500 dados de input, a precisão no conjunto de teste e trainamento cai. Esse fator ocorre pois com apenas um classificador o modelo não consegue aprender mais com novos dados e aumentar o número de dados não favorece muito o resultado final. Isso caracteriza o cenário no qual o modelo tem High bias, pois não aprender mais características ao ser treinado com novos dados.
- **10 Decision Trees**: Analisando a curva de aprendizado para este modelo, é possível observar no começo do eixo x, onde temos pouco treinamento, a precisão em treinamento enquanto a precisão de validação é baixa. A medida que aumentamos o conjunto de teste, a precisão de ambas as linhas convergem. Esse cenário apresenta a situação na qual encontramos low bias e low variance, que é a situação ideal para a construção de um modelo.
- **50 Decision Trees**: Com 50 árvores de decisão, percebemos que a precisão se mantém alta, igual nos outros experimentos, porém neste cenário há um diferença na curva de aprendizado. Quando a capacidade ultrapassa o ponto de 2500 dados de input, percebemos que a precisão em no conjunto de teste diminui, o que caracteriza que o modelo está sendo sobretreinado e que o treinamento deveria ter parado de forma antecipada.
- **100 Decision Trees**: Com 100 decision Trees, percebemos que o modelo se comporta de forma similar à quando utilizamos apenas 10 decision trees. É perceptível que a precisão no conjunto de teste e treinamento convergem, o que caracteriza que o modelo ainda pode ser treinado com mais dados.

## Naive Bayes
O algoritmo Naive Bayes é um classificador que utiliza o teorema de Bayes para conseguir descobrir quais features são independentes de outras. A partir disso o classificador consegue discriminar quais features influenciam no resultado final de classificação.

```python
nb_classifiers = [
    ['Gaussian Naive Bayes', GaussianNB()],
]

trainAndEvaluate(nb_classifiers)
```

Neste experimento único, podemos avaliar a performance do algoritmo a partir dos resultados exibidos nos gráficos acima. A partir da curva de aprendizado, percebemos que o algoritmo possui High Bias, pois a medida que sua capacidade aumenta, sua precisão diminui. \
A partir da matriz de confusão e da curva ROC, percebemos que o Recall do algoritmo possui um score bom (0.97), porém a performance geral do modelo é afetada pelo valor de Specificity menor (0.88). \
Uma suposição do que pode ter feito a precisão do modelo cair a medida que a capacidade aumenta, é uma maior proporção de dados que são classificados como negativo. Isso pode ter abalado a performance geral do modelo pelo fato dele não conseguir classificar estes valores tão bem.

## Decision Tree
A árvore de Decisão é um algoritmo que a cada nó posiciona uma pergunta sobre as features, e desta forma vai filtrando os dados até chegar nas folhas, onde o resultado da classificação estará presente.

```python
dt_classifiers = [
    ['Max Depth 5 Decision Tree', DecisionTreeClassifier(max_depth=5)],
    ['Max Depth 20 Decision Tree', DecisionTreeClassifier(max_depth=20)],
    ['Max Depth 50 Decision Tree', DecisionTreeClassifier(max_depth=50)],
    ['No limit Depth Decision Tree', DecisionTreeClassifier()],
]

trainAndEvaluate(dt_classifiers)
```

Para analisar a performance das Decision Trees, criamos 4 modelos que variam a profundidade máxima das árvores. Estes modelos são:
- **Profundidade Máxima = 5:** A partir da curva de aprendizado deste modelo, podemos perceber que a precisão no conjunto de teste e treinamento convergem à medida que a capacidade aumenta. Quando isso ocorre, podemos dizer que o modelo está sendo treinado corretamento, pois apresenta low bias e low variance.
- **Profundidade Máxima = 20:** Com o valor de 20 para a profundidade máxima, observamos que o score no conjunto de treinamento é 1, ou se aproxima de 1 para qualquer valor de capacidade. À partir de 2500 dados de input, é possível perceber que o score em conjunto de teste começa a cair, isso ocorre pois o modelo está com High variance, onde é perceptível que ele está com Overfit.
- **Profundidade Máxima = 50:** Com a profundidade máxima definida como 50, podemos ver na curva de aprendizado que o score de treinamento ainda é 1 (erro empírico 0), e que a precisão em conjunto de treinamento cresce pouco à medida que a capacidade aumenta. Isso pode indicar que a performance do algoritmo não irá melhorar muito a medida que fornecemos mais dados de treino e a melhor ação a ser feita é parar o treinamento.
- **Sem Limite de Profundidade Máxima:** Observando o gráfico de curva de precisão, percebemos que a situação deste modelo é bem parecidade com o modelo de profundidade 20, onde a capacidade aumenta e o erro esperado aumenta. À partir disso podemos concluir que este modelo está com Overfitting.

## k-NN
O classificador K-nearest Neighbors utiliza a estratégia de encontrar vizinhos próximos de dado em dado para conseguir traçar um separador entre classes diferentes.

```python
k_nn_classifiers = [
    ['3 neighbors Classifier', KNeighborsClassifier(3)],
    ['5 neighbors Classifier', KNeighborsClassifier(5)],
    ['20 neighbors Classifier', KNeighborsClassifier(20)],
    ['50 neighbors Classifier', KNeighborsClassifier(50)],
]

trainAndEvaluate(k_nn_classifiers)
```

A análise deste experimento será feita de forma conjunta, pois os resultados obtidos nos diferentes experimentos possuem características similares. \
De forma geral, conseguimos perceber que todos os modelos possuem um valor alto de Recall (de 0.94 à 0.97), enquanto o valor do Specificity é menor. Como o kNN busca cada item por proximidade, o aspecto citado pode indicar que os valores positivos estão concentrados para determinados valores que as features assumem, enquanto os valores negativos estão mais esparsos. Analisando a curva de aprendizado dos modelos, percebemos que tanto a precisão em conjunto de teste quanto de treinameto aumenta à medida que aumentamos a capacidade, o que indica que o modelo pode ser treinado com mais dados.

## Gradient Tree Boosting
O Gradient Tree Boosting cria uma função de perda e agrega o valor de classificadores mais fracos (decision trees) para construir uma classificação mais robusta.

```python
gtb_classifiers = [
    ['5 iteration GTB', GradientBoostingClassifier(n_iter_no_change=5)],
    ['10 iteration GTB', GradientBoostingClassifier(n_iter_no_change=10)],
    ['20 iteration GTB', GradientBoostingClassifier(n_iter_no_change=20)],
]

trainAndEvaluate(gtb_classifiers)
```

Para avaliar a performance dos modelos de Gradient Tree Boosting, foram criados 3 modelos que variam o valor de interações no treinamento. Podemos perceber que a precisão se manteve por volta de 96.7% em todos os experimentos. \
À partir do valor de Recall, Specificity e dá curva ROC, conseguimos ver que esse classificador tem alta assertividade em prever classes negativas e positivas corretamente. \
À partir das curva de aprendizado, que são similares, percebemos que a precisão no conjunto de teste e treinamento convergem, isso significa que os modelos podem receber mais capacidade, porém a precisão não aumentará muito.
